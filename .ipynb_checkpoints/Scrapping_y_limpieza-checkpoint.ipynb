```python
#Función para tener ordenadas las listas de listas scrapeadas.
def listas_libros(libros):
    lista_libros=[]
    for i in range(4):
        lista = [libro[i] for libro in libros]
        lista_libros.append(lista)
    return lista_libros

scrap_limpio = listas_libros(libro2)
```

```python
#Función para limpiar los votos de los usuarios.
def limpieza_votos(votos):
    a = votos.split(" ")[0]
    if a == "No":
        return None
    else:
        return float(a.replace(",","").replace(".",""))
    
lista_votos_limpio = [limpieza_votos(voto) for voto in scrap_limpio[2]]
```

```python
#Función para limpiar las notas promedio de los libros.
def promedio(rating):
    if rating == "No tiene nota media":
        return None
    else:
        return float(rating)
    
nota_promedio_limpio = [promedio(rati) for rati in scrap_limpio[1]]
```

```python
#Función para limpiar los autores.
def Autor(au):
    if au == "No tiene autor":
        return None
    else:
        return au

Autor_limpio = [Autor(auto) for auto in scrap_limpio[0]]
```

```python
#Fuzzy_wuzzy para filtrar las peliculas que coinciden con los libros.
fuzzy_titulo = []
for (a,b) in zip(scrap_limpio[3],list(df_movies["Series_Title"])):
    porcentaje = fuzz.ratio(a,b)
    if porcentaje > 45:
        porcentaje = True
        fuzzy_titulo.append(porcentaje)
    else:
        porcentaje = False
        fuzzy_titulo.append(porcentaje)
```

```python
df_movies_limpio["Gross"] = df_movies_limpio["Gross"].str.replace(',','')
df_movies_limpio
```

```python
df_movies["Existe_libro"] = fuzzy_titulo
df_movies["Autor_libro"] = Autor_limpio
df_movies["Nota_promedio_libro"] = nota_promedio_limpio
df_movies["Votos_libro"] = lista_votos_limpio
```

```python
#Aqui creo un DF nuevo con solo las filas que coincide que las peliculas tienen libro.
df_movies_limpio = df_movies[df_movies["Existe_libro"] == True]
```
